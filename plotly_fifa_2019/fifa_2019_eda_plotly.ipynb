```python
import pandas as pd
from pprint import pprint
```

```python
df = pd.read_csv("data/fifa_2019_players.csv")
df.describe()
```

```python
list(df.columns)
```

```python
df['Age'].iplot(kind='hist', xTitle='Age', yTitle='count', title='Age Distribution')
```

```python
df[['Finishing', 'Penalties']].iplot(
    kind='hist', barmode='overlay', histnorm='percent', 
    xTitle='Goal scoring attributes', yTitle='(%) of players',
    title='Comparison of goal-scoring methods')
```

Ability to finish is more widely distributed as compared to scoring penalties.
In other words, people are generally better at penalties than finishing. No surprises there.

```python
count_by_country = df.Nationality.value_counts()
top_countries_by_count = list(count_by_country.iloc[0:20,].index)
count_by_country
```

```python
df_overall_by_countries = df[['Nationality', 'Overall']].groupby('Nationality').agg(['mean', 'median', 'max'])
df_overall_by_countries.columns = df_overall_by_countries.columns.get_level_values(1)
df_overall_by_countries
```

```python
df_overall_by_countries_2 = df_overall_by_countries.loc[df_overall_by_countries.index.isin(top_countries_by_count[:10])].or
df_overall_by_countries_2
```

```python
df_overall_by_countries_2.iplot(kind='bar', xTitle='Nationality', yTitle='Average overall score',
                               title='Average overall by country')
```
